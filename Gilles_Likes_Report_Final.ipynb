## The Process ##
Firstly, I needed to import the proper Python library, pandas.

````python
import pandas as pd
import json
````

I then located where that downloaded files lived, and assign it to a variable. With that variable, I then loaded the .json and gave it a name, "my_data."

````python
with open(r'C:\Users\natal\Desktop\DataEMAT social medias\Instagram\likes\liked_posts.json') as j:
    my_data = json.load(j)
````

Next, within this data, I want to go within the ```likes_media_likes``` array to dive deeper into the keys provided.

Using the ```Dataframe()``` function, I can then create a data frame to see how my data is organized. This will be called ```insta_likes```.

````python
insta_likes = pd.DataFrame(my_data['likes_media_likes'])
insta_likes
````

My next task is to get a new column in this data frame to represent the timestamps. I need to confirm with myself how to one timeframe out specifically so I can then do it properly when making a list.

````python
my_data['likes_media_likes'][0]['string_list_data'][0]['timestamp']
````

I am now going to assign the timestamps to ```timestamp_list``` and use ```for x in``` to create a loop which continuously pulls the timeframe from each array.

````python
timestamp_list = [x['string_list_data'][0]['timestamp'] for x in my_data['likes_media_likes']]
````

Next is to add that list to the dataframe, under the name "time."

````python
insta_likes['time'] = timestamp_list
insta_likes
````

The last part is to count how many ```string_list_data``` values there are per account, or title, which will represent how many likes were given per account, and set it to ```ascending = False``` to organize by highest quantity first. Giving it 30 outputs on the data frame will make for a longer list until it reaches a stagnant repeating number per account.

````python
title_group_sort = insta_likes.groupby('title').count().sort_values('string_list_data', ascending = False)
title_group_sort.head(31)
````
